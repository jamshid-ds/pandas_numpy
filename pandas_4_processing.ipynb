```python
df=pd.read_csv("https://raw.githubusercontent.com/MansurCompAI/Practicum_Datasets/master/world_cities_air_water_quality.csv")
df
```

```python
df.info()
```

```python
index=["City","Region","Country","airQuality","WaterPollution"]
df.columns = index
df
```

```python
df["Region"] = df["Region"].map(lambda x: x.replace('"',""))
df["Country"] = df["Country"].map(lambda x: x.replace('"',""))
df["Country"] = df["Country"].map(lambda x: x.replace(' ',""))

df
```

```python
df[df['Country']=="Uzbekistan"]
```

```python
df1 = pd.pivot_table(df,index = ['Country','City'] )
df1
```

```python
df1.loc[['Uzbekistan',"Germany",'Tajikistan']]
```

```python
df2 = df1.copy()
df2
```

```python
df2['AirPollution'] = 100-df2['airQuality']
df2['WaterQuality'] = 100-df2["WaterPollution"]
df2
```

```python
df2.swaplevel("Country",'City')
```

```python
df2.sort_index(level=0,ascending=False)
```

```python
df2.to_csv("sorted.csv")
```
